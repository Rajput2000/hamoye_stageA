## Answers to Quiz questions

#### 11. Total number of Animal Fat produced in 2014 and 2017

```python
df.groupby('Item').sum().loc['Animal fats',['Y2014', 'Y2017']]
```

#### 12. Mean and Standard Deviation of Year 2015 in 2 decimal places

```python
mean = np.round(df['Y2015'].mean(),3)
std = np.round(df['Y2015'].std(), 3)
print(mean,' and ',std)
```

#### 13. Total number and percentage of missing data in year 2016

```python
missing_data = df['Y2016'].isnull().sum()
percent = np.round((missing_data/ df.shape[0]) * 100, 2)
print(missing_data, ' and ',percent)
```

#### 14. Year 2014 has the highest correlation of Years with Element code

```python
df.corr().iloc[2,3:].sort_values(ascending=False)
```

#### 15. Year 2017 has the highest sum of Import Quantity

```python
df.groupby('Element').sum().iloc[7, 3:].sort_values(ascending=False)
```

#### 16. Total number of sum of Production in Year 2014

```python
df.groupby('Element').sum().loc['Production', 'Y2014']
```

#### 17. Element Domestic supply quantity has the highest sum in Year 2018

```python
df.groupby('Element').sum()['Y2018'].sort_values(ascending=False)
```

#### 18. Element Protein supply quantity has the 3rd lowest sum in Year 2018

```python
df.groupby('Element').sum()['Y2018'].sort_values(ascending=True)
```

#### 19. Total import quantity in Algeria in 2018

```python
df.groupby(['Area' , 'Element']).sum().iloc[7]['Y2018']
```

#### 20. Total number of unique countries 

```python
df['Area'].nunique()
```
